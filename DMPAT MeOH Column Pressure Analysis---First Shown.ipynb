```python
## cutting short the df to analyse

df1=df[['C-14601A MEOH FIC REFLUX PV', 'C14601A PRES Bottom. PV']]

clf1= df1.values

# transforming features to analyse better 
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
clf1_scaled = scaler.fit_transform(clf1)
df1_scaled= pd.DataFrame(clf1_scaled)

#converting names
df1_scaled.columns=['C-14601A MEOH FIC REFLUX PV', 'C14601A PRES Bottom. PV']
df1_scaled.index=df.index

df1_scaled.head()
```

```python
plt.figure(figsize=(16, 8), dpi=150)
  

df1_scaled['C-14601A MEOH FIC REFLUX PV'].plot(label='C14601A Reflux', color='orange')
df1_scaled['C14601A PRES Bottom. PV'].plot(label='C14601A Bottom Pressure')
plt.title('Comparison Plot')
  
# adding Label to the x-axis
plt.xlabel('Date and Time')

plt.grid()
  
# adding legend to the curve
plt.legend()
```
